# Deep Learning on MNIST Dataset

```python
import numpy as np
import pandas as pd 
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
```

### load dataset and study it

```python
(x_train,y_train),(x_test,y_test) = keras.datasets.mnist.load_data()
```

```python
print(x_train)
print(y_train)
print(x_train.shape)
print(y_train.shape)
```

### check out data in x_rain and y_train and test set too

```python
# data contains values ranging from 0 to 255
# There are 60k records in x_train each record having size 28*28

print(x_train[0].shape)
print(x_train[0])

print(y_train[0].shape)
print(y_train[0])

```

```python
print(x_test.shape)
print(x_test)
print(y_test.shape)
print(y_test)
```

### Data Scalling
 data scalling is needed as data values are ranging from 0 to 255
 itmay affect the performance of the model

```python
x_train = x_train/255
x_test = x_test/255
```

```python
x_train[0]
```

### convert 3 d array to 2 d array


```python
x_train_flat = x_train.reshape(len(x_train), 28*28)
x_test_flat = x_test.reshape(len(x_test), 28*28)
```

```python
print(x_train_flat.shape)
print(x_test_flat.shape)
```

### Build Model

```python
keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,),activation='sigmoid')])

keras.Model().compile(optimizer='rmsprop',
    loss= "categorical_crossentropy",
    metrics= ["accuracy"])
```
